```python
import pandas as pd
```

```python
def df2qs(df:pd.DataFrame):
    qs = {}
    for i, row in df.iterrows():
        q = row['quality_metrics']
        for n in q:
            if n in qs:
                qs[n].append(q[n])
            else:
                qs[n] = [q[n]]

    return qs
```

```python
import matplotlib.pyplot as plt


def plot_qs(qs):
    plt.figure(dpi=300, facecolor="white")
    plt.subplots_adjust(wspace=0.5, hspace=0.8)

    placement_i = 1

    for name in qs:
        plt.suptitle(f"{len(qs[name])}")
        plt.subplot(2, 5, placement_i)
        plt.xticks(fontsize=5, rotation=-45)
        plt.yticks(fontsize=5)
        plt.title(f"{name}", fontsize=5)
        plt.boxplot(qs[name], whis=100)
        placement_i += 1

    plt.show()


def plot_qs_comp(qss):
    plt.figure(dpi=300, facecolor="white")
    plt.subplots_adjust(wspace=0.5, hspace=0.8)

    placement_i = 1

    for name in qss[0]:
        plt.suptitle(f"{len(qss[0][name])}")
        plt.subplot(2, 5, placement_i)
        plt.xticks(fontsize=5, rotation=-45)
        plt.yticks(fontsize=5)
        plt.title(f"{name}", fontsize=5)
        plt.boxplot([qs[name] for qs in qss], whis=100)
        placement_i += 1

    plt.show()
```
