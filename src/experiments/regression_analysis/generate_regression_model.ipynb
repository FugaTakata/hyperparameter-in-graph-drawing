# 3つのグラフのデータを利用して、線形回帰関数を作成する


```python
from config.paths import root_path
import pandas as pd
```

```python
dataset_names = ["les_miserables", "1138_bus", "USpowerGrid"]
params_names = sorted(["eps", "number_of_iterations", "number_of_pivots"])
```

```python
grid = pd.DataFrame()
for dataset_name in dataset_names:
    grid_data_path = root_path.joinpath(
        f"data/experiments/regression_analysis/grid/{dataset_name}.pkl"
    )
    grid = pd.concat([grid, pd.read_pickle(grid_data_path)])
```

```python
grid.query('dataset_name == "USpowerGrid" & number_of_pivots == 50 & number_of_iterations == 10 & eps == 0.05')
```

```python
from sklearn.model_selection import KFold
from itertools import combinations

params_values = {}
for p_name in params_names:
    params_values[p_name] = sorted(list(set(grid[p_name].values)))

p_combs = list(combinations(params_names, 2))
for p_comb in p_combs:
    fp = p_comb[0:2]
    nfp = list(filter(lambda x: x not in fp, params_names))
    print(fp, nfp)
    break
```
