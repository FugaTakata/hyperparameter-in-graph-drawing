```python
import json
import os
import decimal
candidates = {
    'number_of_pivots_rate': [],
    'number_of_iterations': [],
    'eps': []
}

# 有効数字4桁指定
decimal.getcontext().prec = 2

for i in range(1, 21):
    a = decimal.Decimal(i) * decimal.Decimal(0.05)
    candidates['number_of_pivots_rate'].append(float(a))


for i in range(1, 11):
    a = 5 * i
    candidates['number_of_iterations'].append(a)

for i in range(6, 11):
    a = 10 * i
    candidates['number_of_iterations'].append(a)

for i in range(3, 11):
    a = 50 * i
    candidates['number_of_iterations'].append(a)

for i in range(6, 11):
    a = 100 * i
    candidates['number_of_iterations'].append(a)


decimal.getcontext().prec = 2

for i in range(1, 11):
    a = decimal.Decimal(0.1) * decimal.Decimal(i)
    candidates['eps'].append(float(a))

for i in range(1, 10):
    a = decimal.Decimal(0.01) * decimal.Decimal(i)
    candidates['eps'].append(float(a))

for i in range(1, 10):
    a = decimal.Decimal(0.001) * decimal.Decimal(i)
    candidates['eps'].append(float(a))

for k in candidates:
    candidates[k] = sorted(candidates[k])


os.makedirs('data/params/search/candidates', exist_ok=True)

with open(f'data/params/search/candidates/{SS}.json', mode='w') as f:
    json.dump(candidates, f, ensure_ascii=False)
```
